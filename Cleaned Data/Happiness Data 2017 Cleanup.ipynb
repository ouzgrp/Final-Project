```python
# Import our dependencies
import pandas as pd
import numpy as np
from pathlib import Path
from collections import Counter
```

```python
# Load the dataset.
file_path = "2017.csv"
happy_df = pd.read_csv(file_path, index_col=0)
happy_df.head(10)
```

```python
happy_df = happy_df.drop(columns=["Dystopia.Residual", "Whisker.high", "Whisker.low"])
happy_df.head(10)
```

```python
happy_df.rename(columns = {'Economy..GDP.per.Capita.':'Economy (GDP per Capita)', 'Happiness.Rank':'Happiness Rank', 'Health..Life.Expectancy.':'Health (Life Expectancy)', 'Trust..Government.Corruption.':'Trust (Government Corruption)', 'Happiness.Score':'Happiness Score'}, inplace = True)
happy_df.head(10)
```

```python
happy_df = happy_df.reindex(columns=['Happiness Rank','Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity'])
happy_df.head(10)
```

```python
happy_df.to_csv('2017_cleaned.csv', index=True)
```
