```python
# Import our dependencies
import pandas as pd
import numpy as np
from pathlib import Path
from collections import Counter
```

```python
# Load the dataset.
file_path = "2019.csv"
happy_df = pd.read_csv(file_path)
happy_df.head(10)
```

```python
happy_df.rename(columns = {'Country or region':'Country','Overall rank':'Happiness Rank','GDP per capita':'Economy (GDP per Capita)','Healthy life expectancy':'Health (Life Expectancy)','Social support':'Family','Perceptions of corruption':'Trust (Government Corruption)','Score':'Happiness Score','Freedom to make life choices':'Freedom'}, inplace = True)
happy_df.head(10)
```

```python
happy_df = happy_df.reindex(columns=['Country','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity'])
happy_df.head(10)
```

```python
 happy_df.to_csv('2019_cleaned.csv', index=False)
```
