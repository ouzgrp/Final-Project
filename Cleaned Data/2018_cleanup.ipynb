```python
#import our dependencies
import pandas as pd
import numpy as np
from pathlib import Path
from collections import Counter

from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import confusion_matrix
from imblearn.metrics import classification_report_imbalanced
```

```python
#Load dataset
file_path = "2018.csv"
happy_df = pd.read_csv(file_path, index_col=False)
happy_df.head(10)
```

```python
#change column names
happy_df.rename(columns = {'Overall rank':'Happiness Rank','Country or region':'Country','GDP per capita':'Economy (GDP per Capita)','Healthy life expectancy':'Health (Life Expectancy)','Social support':'Family','Perceptions of corruption':'Trust (Government Corruption)','Score':'Happiness Score','Freedom to make life choices':'Freedom'}, inplace = True)
happy_df.head(10)
```

```python
#re-arrange columns
happy_df = happy_df.reindex(columns=['Country','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity'])
happy_df.head(10)
```

```python
happy_df.to_csv('2018_cleaned.csv', index=False)
```
